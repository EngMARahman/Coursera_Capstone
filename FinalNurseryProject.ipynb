```python
# extract the coordinates of 3 search location centers, west, center and east of Cairo
urls=[]
res=[]
for i, loc in enumerate(['Heliopolis Cairo Egypt', 'Maadi Cairo Egypt', 'Giza Egypt']):
    #address = loc
    geolocator = Nominatim()
    lc = geolocator.geocode(loc)
    lat = lc.latitude
    long = lc.longitude
    search_query_1 = 'british school' #international keyword reflects many results, so british is used instead
    search_query_2 = 'mall' 
    search_query_3 = 'business'
    search_query_4 = 'nursery'
    r1 = 3000 # Search within 3km radius from each area center
    r2 = 10000 # Search within 3km radius from each area center
    r3 = 10000 # Search within 3km radius from each area center
    r4 = 10000 # Search within 3km radius from each area center
    
    
    url1 = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'\
            .format(CLIENT_ID, CLIENT_SECRET, lat, long, VERSION, search_query_1, r1, LIMIT)
    url2 = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'\
            .format(CLIENT_ID, CLIENT_SECRET, lat, long, VERSION, search_query_2, r2, LIMIT)
    url3 = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'\
            .format(CLIENT_ID, CLIENT_SECRET, lat, long, VERSION, search_query_3, r3, LIMIT)
    url4 = 'https://api.foursquare.com/v2/venues/search?client_id={}&client_secret={}&ll={},{}&v={}&query={}&radius={}&limit={}'\
            .format(CLIENT_ID, CLIENT_SECRET, lat, long, VERSION, search_query_4, r4, LIMIT)
    
    res_schools=requests.get(url1).json()
    res_malls=requests.get(url2).json()
    res_business=requests.get(url3).json()
    res_nursery=requests.get(url4).json()
    
    urls.append([url1,url2,url3,url4]) 
    res.append([res_schools,res_malls,res_business,res_nursery])
```
